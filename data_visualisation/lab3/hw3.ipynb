```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from math import pi
```

```python
file_path = '/home/ubuntu/year3/data_visualisation/Homework_sampledata1.txt'

# Open the file and process the content
with open(file_path, 'r') as file:
    data_lines = [
        [float(value) for value in line.split()[1:]]  
        for line in file.readlines()[1:]  
    ]

data_lines
```

```python
first_day_values = data_lines[0]

first_day_dataframe = pd.DataFrame([
    {
        'Date': f'2008-{month_index + 1:02}-01',  
        'Time': '00:00:00',  
        'Temperature': None if temp == -99 else temp 
    }
    for month_index, temp in enumerate(first_day_values)
])


first_day_dataframe
```

```python

num_days = len(data_lines)
num_months = len(data_lines[0])

all_days_data = [
    {
        'Date': f'2008-{month_index + 1:02}-{day_index + 1:02}',  
        'Time': '00:00:00',  
        'Temperature': None if data_lines[day_index][month_index] == -99 else data_lines[day_index][month_index]  
    }
    for month_index in range(num_months)
    for day_index in range(num_days)
]

all_days_df = pd.DataFrame(all_days_data)
all_days_df.to_csv('all_days_data.csv', index=False)
print("Data saved to all_days_data.csv")
```

```python
all_days_df['Date'] = pd.to_datetime(all_days_df['Date'], errors='coerce')
all_days_df['Month'] = all_days_df['Date'].dt.month
```

```python

data = np.array(data_lines).T
data = np.where(data == -99.0, np.nan, data)

plt.figure(figsize=(10, 6))
plt.contourf(data, cmap='jet', levels=100)
plt.title('Temperature Contour Plot for 2008')
plt.colorbar(label='Temperature')
plt.xlabel('Day')
plt.ylabel('Month')
plt.yticks(np.arange(12), np.arange(1, 13)) 
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(data, cmap='jet', cbar_kws={'label': 'Temperature'}, xticklabels=2)
plt.title('Temperature Heatmap for 2008')
plt.xlabel('Day')
plt.ylabel('Month')

plt.yticks(np.arange(12), np.arange(1, 13)) 
plt.gca().invert_yaxis()  

plt.show()
```

```python
mean = np.array([
    [np.nanmean(data[i, j*5:j*5 + (5 if j < 5 else 6)]) for j in range(6)]
    for i in range(12)
])

no_category = mean.shape[1]
list_month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

fig, axs = plt.subplots(3, 4, figsize=(13, 7), subplot_kw=dict(polar=True))
plt.suptitle('Radar chart of average temperature in 2008', size=16)

for i, ax in enumerate(axs.flatten()):
    angles = [n / float(no_category) * 2 * pi for n in range(no_category)]
    angles += angles[:1] 
    data_mean = np.concatenate((mean[i, :], mean[i, :1]))

    ax.plot(angles, data_mean, linewidth=1, linestyle='solid')
    ax.fill(angles, data_mean, 'r', alpha=0.1)
    
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(['c1', 'c2', 'c3', 'c4', 'c5', 'c6'], color='grey', size=8)
    ax.set_yticks([10, 20, 30])
    ax.set_yticklabels(["10", "20", "30"], color="grey", size=7)
    ax.set_ylim(0, 40)
    ax.set_title(list_month[i], size=15, y=-0.3)

plt.subplots_adjust(wspace=0.5, hspace=0.6)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

link_mean = linkage(mean, 'ward')
plt.figure(figsize=(12, 5))
dendrogram(link_mean, labels=list_month)
plt.title('Dendrogram of average temperature in 2008', size=16)
plt.show()
```
